```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def median_blur_manual(img, ksize=3):
    """
    ksize: kernel size (must be odd: 3, 5, 7, ...)
    """
    pad = ksize // 2
    # Pad image to handle borders
    padded = np.pad(img, pad, mode="edge")
    output = np.zeros_like(img)

    for i in range(img.shape[0]):
        for j in range(img.shape[1]):
            # Region
            region = padded[i:i+ksize, j:j+ksize]
            # Median of the region
            median_val = np.median(region)
            output[i, j] = median_val

    return output

noisy_img = cv2.imread("noisy_charlie_chaplin.png", cv2.IMREAD_GRAYSCALE)

# Manual mendian filter
median_3_manual = median_blur_manual(noisy_img, 3)
median_5_manual = median_blur_manual(noisy_img, 5)
median_7_manual = median_blur_manual(noisy_img, 7)

# OpenCV version
median_3_cv = cv2.medianBlur(noisy_img, 3)
median_5_cv = cv2.medianBlur(noisy_img, 5)
median_7_cv = cv2.medianBlur(noisy_img, 7)

# Show results
plt.figure(figsize=(15,8))

plt.subplot(2,4,1)
plt.title("Original (Noisy)")
plt.imshow(noisy_img, cmap="gray")
plt.axis("off")

plt.subplot(2,4,2)
plt.title("Manual k=3")
plt.imshow(median_3_manual, cmap="gray")
plt.axis("off")

plt.subplot(2,4,3)
plt.title("Manual k=5")
plt.imshow(median_5_manual, cmap="gray")
plt.axis("off")

plt.subplot(2,4,4)
plt.title("Manual k=7")
plt.imshow(median_7_manual, cmap="gray")
plt.axis("off")

plt.subplot(2,4,6)
plt.title("OpenCV k=3")
plt.imshow(median_3_cv, cmap="gray")
plt.axis("off")

plt.subplot(2,4,7)
plt.title("OpenCV k=5")
plt.imshow(median_5_cv, cmap="gray")
plt.axis("off")

plt.subplot(2,4,8)
plt.title("OpenCV k=7")
plt.imshow(median_7_cv, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
