```python
import cv2
import matplotlib.pyplot as plt

# Noisy Image
noisy_img = cv2.imread("noisy_charlie_chaplin.png", cv2.IMREAD_GRAYSCALE)

# Different kernel size for comparison
median_3 = cv2.medianBlur(noisy_img, 3)
median_5 = cv2.medianBlur(noisy_img, 5)
median_7 = cv2.medianBlur(noisy_img, 7)

# Show results
plt.figure(figsize=(12,6))

plt.subplot(1,4,1)
plt.title("Original (Noisy)")
plt.imshow(noisy_img, cmap="gray")
plt.axis("off")

plt.subplot(1,4,2)
plt.title("Median k=3")
plt.imshow(median_3, cmap="gray")
plt.axis("off")

plt.subplot(1,4,3)
plt.title("Median k=5")
plt.imshow(median_5, cmap="gray")
plt.axis("off")

plt.subplot(1,4,4)
plt.title("Median k=7")
plt.imshow(median_7, cmap="gray")
plt.axis("off")

plt.show()
```
